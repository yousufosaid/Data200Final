## Project Goals

- Predict player scoring (`PTS`) and player efficiency (`PER`) with interpretable regression models (linear + regularized + non-linear variants).
- Classify player position groups (Guard vs Big) and optionally team playoff qualification using classification models and clear evaluation metrics.
- Discover player archetypes using hierarchical clustering and K-means and interpret clusters with feature summaries.

## Data

Repository-relative files used:
- player_basic.csv — season basic per-player stats (PTS, AST, TRB, FG, 3P, MP, G, etc.)
- player_advanced.csv — advanced per-player metrics (PER, TS%, USG%, VORP, WS, BPM, etc.)
- nbateamdata.csv — team-level statistics (PTS per game, FG/FGA, TRB, AST, etc.)

Each notebook will include a short Data Description cell showing shapes, dtypes, missing values, and a small sample of rows.

## Methods & Workflow

- Data cleaning and merging using `pandas` (imputation, removal of aggregated rows like 'League Average').
- Feature engineering: create at least five engineered features (mix of continuous, discrete, categorical). Examples: `3P_rate = 3P/3PA`, `efficiency = PTS/MP`, `minutes_bin`, `experience_years`, `position_group`.
- Modeling: use sklearn `Pipeline` to prevent leakage; perform CV and hyperparameter tuning (`GridSearchCV` / `LassoCV`).
- Regression: `LinearRegression` baseline, `Ridge`/`Lasso` with CV, plus a tree-based regressor (RandomForest/GradientBoosting). Evaluate with RMSE and R².
- Classification: `LogisticRegression` baseline (tune regularization) and a tree-based classifier; evaluate using accuracy, precision/recall, confusion matrix, and ROC/AUC.
- Clustering: hierarchical clustering and K-means, report silhouette scores and cluster summaries.
- Interpretability: coefficients for linear models, feature importances for tree models; optional SHAP for deeper explanation.

## Project Conclusion

This project analyzed player- and team-level NBA statistics to (1) predict scoring and efficiency, (2) classify player roles, and (3) discover statistical player archetypes. Below are the main results, interpretation, limitations, and final recommendations.

**Main results:**
- Regression (PTS): A linear modeling approach with Lasso selection produced a highly predictive, interpretable model — non-scoring predictors such as `VORP` and efficiency metrics explain a large portion of scoring variance after removing direct scoring variables. Lasso reduced the predictor set and highlighted valuable features for interpretation.
- Regression (PER): Predicting PER from basic stats improved when selective polynomial terms were included; `eFG%` and `TOV` consistently show strong influence on PER estimates.
- Classification: Logistic regression on advanced metrics separates `Guard` vs `Big` reasonably well (advanced stats outperform raw volume), and hierarchical/K-means clustering reveals coherent player groups that match intuitive role differences.

**Interpretation & takeaway:**
- Efficiency and value-over-replacement measures (e.g., `eFG%`, `VORP`, `PER`) are more informative for distinguishing high-impact players than raw volume alone. Regularized linear models give transparent, actionable coefficients for interpretation.
- Clustering complements supervised models by highlighting archetypes (shooters, playmakers, rebound/defensive specialists) that can guide feature construction and team-level strategies.

**Limitations:**
- Some preprocessing and modelling steps currently exist as ad-hoc cells; to ensure reproducibility and avoid leakage these should be moved into `Pipeline` objects and CV workflows.
- Tree-based and ensemble models (Random Forest / Gradient Boosting) were not yet used for final comparison — they may improve predictive accuracy at the cost of interpretability.
- The notebooks use season-aggregate stats; per-game or possession-normalized features would reduce bias from playing time and team pace.
